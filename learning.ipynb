```python
degrees = np.array([90, 180, 270, 360])
radians = np.deg2rad(degrees)
radians = np.round(radians, 3)
print(radians)
x = np.sin(radians)
x = np.round(x, 2)
print(x)
y = np.round(np.arcsin(x), 3)
print(y)
```

```python
degrees = np.array([0, 90, 180, 270, 360])
radians = np.deg2rad(degrees)
radians = np.round(radians, 3)
print(radians)
x = np.cos(radians)
x = np.round(x, 2)
print(x)
y = np.round(np.arccos(x), 3)
print(y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,4,6,8,10])

plt.plot(x,y)

model = LinearRegression()
model.fit(x, y)

prediction = model.predict([[6]])
print(prediction)
```
